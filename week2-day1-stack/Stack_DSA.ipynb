## **âœ… 2. Use a Python Singly LinkedList**

```python
from typing import ItemsView
class Node:
  def __init__(self,item=None, next=None):
    self.item = item
    self.next = next

class stack:
  def __init__(self):
    self.top = None
    self.item_count = 0

  def is_empty(self):
    return self.top == None

# PUSH OPERATION:
  def push(self, data):
    n = Node(data, self.top)
    self.top = n
    self.item_count += 1

# POP OPERATION:
  def pop(self):
    if not self.is_empty():
       data = self.top.item
       self.top = self.top.next  # Corrected from self.start
       self.item_count -= 1
       return data
    else:
      raise IndexError ("stack is empty")

# PEEK OPERATION:
  def peek(self):
    if not self.is_empty():
      return self.top.item
    else:
      raise IndexError("stack is empty") # Corrected error handling

# SIZE OPERATION:
  def size(self):
    return self.item_count
```

## **Creation of Stack:** ðŸ“¦

```python
# STACK OBJECT CREATION
s1 = stack()
s1.push(10)
s1.push(20)
s1.push(30)

print("Total elements in the stack=", s1.size())
print("Top elements in the stack=", s1.peek())
print("Poped elements in the stack=", s1.pop())


#again, we can see the modified output in the stack
print("Total elements in the stack=", s1.size())
print("Top elements in the stack=", s1.peek())
```
